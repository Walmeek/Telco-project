```python
user_behaviour=data.groupby("Bearer Id").agg({'Dur. (ms)': "count",
                                           'Start ms': 'min',
                                           'End ms': 'max',
                                           'Total UL (Bytes)': 'sum',
                                           'Total DL (Bytes)': 'sum',
                                           'Other UL (Bytes)': 'sum',
                                           'Other DL (Bytes)': 'sum',
                                           'Gaming UL (Bytes)': 'sum',
                                           'Gaming DL (Bytes)': 'sum',
                                           'Netflix UL (Bytes)': 'sum',
                                           'Netflix DL (Bytes)': 'sum',
                                           'Youtube UL (Bytes)': 'sum',
                                           'Youtube DL (Bytes)': 'sum',
                                           'Email UL (Bytes)': 'sum',
                                           'Email DL (Bytes)': 'sum',
                                           'Google UL (Bytes)': 'sum',
                                           'Google DL (Bytes)': 'sum',
                                           'Social Media UL (Bytes)': 'sum',
                                           'Social Media DL (Bytes)': 'sum'
                                    
                                           
                                           }).reset_index()
print(user_behaviour)
```
